```python
# quantiles

q1=df['Population in thousands (2017)'].quantile(0.25)
q2=df['Population in thousands (2017)'].quantile(0.5)
q3=df['Population in thousands (2017)'].quantile(0.75)
```

```python
print(q1,q2,q3)
```

```python
#iqr

iqr=q3-q1
iqr
```

```python
# lower fence and upper fence

lf=q1-1.5*iqr
uf=q3+1.5*iqr
```

```python
print(lf,uf)
```
